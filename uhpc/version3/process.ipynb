```python
import pandas as pd
df_cs = pd.read_csv('Compressive_strength.csv').astype(float)
df_fs = pd.read_csv('Flexural_strength.csv').astype(float)
df_p = pd.read_csv('Porosity.csv').astype(float)
df_ss = pd.read_csv('Mini_slump_spread.csv').astype(float)
```

```python
cols_list_cs = list(df_cs.columns)
cols_list_fs = list(df_fs.columns)

```

```python
# 判断两个列表是否相同，如果不同，则输出不同的列名
if cols_list_cs != cols_list_fs:
    print("不同的列名:")
    print(set(cols_list_cs) ^ set(cols_list_fs))
else:
    print("列名相同")
```

```python
# 删除最后两个列名
cols_list_cs.pop()
cols_list_cs.pop()

cols_list_fs.pop()
cols_list_fs.pop()

```

```python
df_final = pd.merge(
    df_cs,
    df_fs,
    on=cols_list_cs,
    how='outer',          # outer → 保留所有样本
    indicator=True
)
```
